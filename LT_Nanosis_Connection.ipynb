```python
# prompt: draw graph
# node,a,b,c
# edge(a,b)



import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.MultiGraph()

# Add nodes
G.add_node("PMD4",purpose = 'piezo motor drive'); G.add_node("Handset",purpose = 'piezo controller by hand')
G.add_node("HVS4",purpose = 'high voltage supplier')
G.add_node("HVA4",purpose = 'high voltage amplifier')
G.add_node("SC4",purpose = 'signal controler')
G.add_node("RC4",purpose = 'real controler')


# Add edge
G.add_edge("PMD4", "HVS4",name='PMD4.SV - HVS4.HO1')
G.add_edge("PMD4", "HVA4",name='PMD4.DPA - HVA4.DPAT')
G.add_edge("PMD4", "Handset",name='PMD4.H - Handset')

G.add_edge("HVS4", "HVA4",name='HVS4.HO2 - HVA4.SV')

G.add_edge("HVA4", "SC4",name='HVA4.DPA - SC4.DPA')


# Draw the graph
nx.draw(G, with_labels=True, node_size=700, node_color="skyblue", font_size=10, width=2)

# Show the plot
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a MultiGraph (undirected graph that allows multiple edges)
G = nx.MultiGraph()

# Add nodes
G.add_nodes_from([1, 2])

# Add multiple edges between the same pair of nodes
G.add_edge(1, 2)
G.add_edge(1, 2)  # Adding a second edge between 1 and 2
G.add_edge(1, 2)  # Adding a third edge between 1 and 2

# Draw the graph with curved edges to distinguish parallel edges
pos = nx.spring_layout(G)
plt.figure(figsize=(6, 4))

# Draw the nodes
nx.draw_networkx_nodes(G, pos, node_size=700)

# Draw the labels
nx.draw_networkx_labels(G, pos, font_size=14)

# Draw the edges with distinct curvature
for i, (u, v, key) in enumerate(G.edges(keys=True)):
    rad = 0.1 * (i - 1)  # Curve different for each edge
    nx.draw_networkx_edges(G, pos, edgelist=[(u, v)],
                           connectionstyle=f'arc3,rad={rad}',
                           edge_color='black')

plt.axis('off')
plt.title("MultiGraph with Multiple Edges between Nodes 1 and 2")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import itertools

# Create a MultiGraph
G = nx.MultiGraph()

# Add nodes
G.add_nodes_from(['A', 'B', 'C', 'D'])

# Add multiple edges between nodes
# Edge 1: A to B, with a default key (0)
G.add_edge('A', 'B', weight=10, type='road')
# Edge 2: A to B, with a new key (1)
G.add_edge('A', 'B', weight=5, type='path')
# Edge 3: A to B, with a new key (2)
G.add_edge('A', 'B', weight=15, type='river')

# Add other edges for context
G.add_edge('B', 'C', weight=7, type='trail')
G.add_edge('C', 'D', weight=20, type='highway')
G.add_edge('D', 'A', weight=8, type='track')

# Also add a self-loop with multiple edges
G.add_edge('A', 'A', weight=2, type='loop1')
G.add_edge('A', 'A', weight=3, type='loop2')

# Define a layout for the nodes
pos = nx.spring_layout(G, seed=42) # For consistent layout

plt.figure(figsize=(10, 8))

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=2000)

# Draw node labels
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

# --- Visualize multiple edges with different connection styles ---
# To avoid overlapping edges, we need to draw them iteratively with different 'rad' values.
# We'll use a dictionary to store 'rad' values for each pair of nodes with multiple edges.
# The 'rad' value determines the curvature of the arc.

# For self-loops, we'll use a larger 'rad' value
self_loop_rad = 0.2

# For multi-edges between two different nodes
# We'll generate a sequence of 'rad' values to spread them out
# Start with a base radian, then increment for each additional edge
base_rad = 0.1
rad_increment = 0.08 # Smaller increment for inter-node edges

drawn_edges = set() # To keep track of edges we've already processed for connectionstyle
edge_paths = {} # Stores { (u, v): [list of connectionstyle strings] }

for u, v, key, data in G.edges(keys=True, data=True):
    # Handle self-loops separately
    if u == v:
        # Assign a distinct curve for each self-loop
        if (u, v) not in edge_paths:
            edge_paths[(u, v)] = []
        edge_paths[(u, v)].append(f"arc3,rad={self_loop_rad * (len(edge_paths[(u, v)]) + 1)}")
    else:
        # For multi-edges between different nodes
        # Sort (u, v) to treat (A, B) and (B, A) the same for consistent styling
        sorted_nodes = tuple(sorted((u, v)))
        if sorted_nodes not in edge_paths:
            edge_paths[sorted_nodes] = []

        # Determine the radian based on the number of edges already processed for this pair
        current_rad = base_rad + (len(edge_paths[sorted_nodes]) * rad_increment)
        # Alternate signs for curvature to draw them on both sides
        current_rad_signed = current_rad * (1 if len(edge_paths[sorted_nodes]) % 2 == 0 else -1)
        edge_paths[sorted_nodes].append(f"arc3,rad={current_rad_signed}")

# Now, iterate through all edges and draw them with their assigned connection style
edge_labels = {}
for i, (u, v, key, data) in enumerate(G.edges(keys=True, data=True)):
    # Determine the connection style for the current edge
    conn_style_idx = 0
    if u == v:
        # Find the specific self-loop's index
        loop_edges = [k for node_u, node_v, k_idx in G.edges(keys=True) if node_u == u and node_v == v]
        conn_style_idx = loop_edges.index(key) # Get the index for the current key
        current_connection_style = edge_paths[(u, v)][conn_style_idx]
    else:
        sorted_nodes = tuple(sorted((u, v)))
        # Find the specific edge's index for this pair
        pair_edges = [(k_idx, d_val) for node_u, node_v, k_idx, d_val in G.edges(keys=True, data=True) if tuple(sorted((node_u, node_v))) == sorted_nodes]
        # Sort them by key to ensure consistent application of connection styles
        pair_edges.sort(key=lambda x: x[0])
        conn_style_idx = [e[0] for e in pair_edges].index(key)
        current_connection_style = edge_paths[sorted_nodes][conn_style_idx]


    nx.draw_networkx_edges(
        G, pos,
        edgelist=[(u, v, key)],  # Draw only this specific edge
        edge_color='gray',
        connectionstyle=current_connection_style,
        alpha=0.7
    )

    # Store edge labels, including the key
    label_text = f"{data['type']} (W:{data['weight']})"
    # For multi-edges, adjust label position slightly to avoid overlap
    label_offset = 0.1 * (conn_style_idx - (len(edge_paths.get(tuple(sorted((u, v))), [])) - 1) / 2) # Adjust offset based on edge count

    # Customize label position for self-loops to prevent crowding
    if u == v:
        label_pos = 0.6 + (conn_style_idx * 0.1)  # Move self-loop labels further out
    else:
        label_pos = 0.5 + label_offset # Default center for other edges

    edge_labels[(u, v, key)] = label_text


nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='darkgreen',
                             bbox={"alpha": 0},  # No background box for labels
                             rotate=False) # Keep labels horizontal

plt.title("NetworkX MultiGraph Visualization with Multiple Edges")
plt.axis('off') # Hide axes
plt.show()
```
