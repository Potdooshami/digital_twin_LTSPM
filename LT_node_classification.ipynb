```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
```

```python
G.add_node("LT-system") #parentNode
```

```python
def birth(G,parent,childs):
  #G is digraph
  #Parent is a node included in G
  #childs are a nodes will be added as child of parent
  for child in childs:
    G.add_edge(parent,child)
  return G
```

```python
G = birth(G,"LT-system",["STM","nanonis","Main box","Sub box","infra"])
G = birth(G,"nanonis",["PMD4","HVS4","HVA4","SC4","RC4","Handset"])
G = birth(G,"Main box",["LakeShore","IonGun","UPS","ErLEED","Korea","IonMain","IonPrep","TSPctrl","HeaterBox"])
G = birth(G,"Sub box",["airBlockCtrl","guageCtrl"])
G = birth(G,"STM",["Vibration","LL","Prep","Main"])
G = birth(G,"infra",["PC","Power","Air"])
G = birth(G,"Main",["head"])
G = birth(G,"head",["P1","P2","P3","P4","P5"])
```

```python

nx.draw(G,with_labels=True)
```
